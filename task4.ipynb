```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

```

```python
# Load the dataset
data = pd.read_csv("C:/Users/user/Downloads/senti.csv")  # Update this path to where your CSV file is located

# Display the first few rows
print(data.head())
```

```python
# Check for missing values and data types
print(data.info())
print(data.describe())
print(data['labels'].value_counts())
```

```python
# Set up the visual style
sns.set(style="whitegrid")

# Plot the distribution of sentiment labels
plt.figure(figsize=(10, 6))
sns.countplot(x='labels', data=data, palette='viridis')
plt.title('Distribution of Sentiment Labels')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# Generate word clouds for each sentiment
sentiments = data['labels'].unique()

for sentiment in sentiments:
    plt.figure(figsize=(10, 6))
    text = ' '.join(data[data['labels'] == sentiment]['tweets'])
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'Word Cloud for {sentiment} Sentiment')
    plt.axis('off')
    plt.show()
```
